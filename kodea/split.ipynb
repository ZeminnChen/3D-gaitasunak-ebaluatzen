```python
import os
from sklearn.model_selection import train_test_split

all_indices = list(range(70000))

# Train: 59 500
# Valid: 10 500
train_idx, val_idx = train_test_split(
    all_indices, 
    test_size=0.15, 
    random_state=42, 
    shuffle=True
)
```

```python
import json
split_data = {'train': train_idx, 'val': val_idx}
with open('../probatzen.json', 'w') as f:
    json.dump(split_data, f)
```
